```python
import sqlite3
```

```python
pitchingRankingQuery = "SELECT team,((SUM(p_er) * 9.0) / ((SUM(p_ipouts) / 3.0))) as \"p_era\",((SUM(p_h) + SUM(p_w)) / (SUM(p_ipouts) / 3.0)) as \"p_whip\",SUM(p_k) from pitching where date like \"%2025%\" GROUP BY team"


teamToEra        = {}
teamToWhip       = {}
teamToStrikeouts = {}

teamEras       = []
teamWhips      = []
teamStrikeouts = []

connection = sqlite3.connect('player_batting.db')
cursor     = connection.cursor()

cursor.execute(pitchingRankingQuery)

rows = cursor.fetchall()

for row in rows:
    team       = row[0]
    era        = float(row[1])
    whip       = float(row[2])
    strikeouts = int  (row[3])

    teamToEra       [team] = era
    teamToWhip      [team] = whip
    teamToStrikeouts[team] = strikeouts

    teamEras          .append(era)
    teamWhips         .append(whip)
    teamStrikeouts    .append(strikeouts)

    print(f"{team}: {era}, {whip}, {strikeouts}")

cursor    .close()
connection.close()    
```

```python
teamEras      .sort()
teamWhips     .sort()
teamStrikeouts.sort(reverse=True)

teamToEraRank       = {}
teamToWhipRank      = {}
teamToStrikeoutRank = {}

for key in teamToEra:
    teamEra        = teamToEra       [key]
    teamWhip       = teamToWhip      [key]
    teamStrikeout  = teamToStrikeouts[key]

    teamEraRank        = teamEras      .index(teamEra       ) + 1
    teamWhipRank       = teamWhips     .index(teamWhip      ) + 1
    teamStrikeoutsRank = teamStrikeouts.index(teamStrikeout ) + 1

    print(f"{key}, {teamEra} : {teamEraRank}, {teamWhip}: {teamWhipRank}, {teamStrikeout}: {teamStrikeoutsRank}")
```
