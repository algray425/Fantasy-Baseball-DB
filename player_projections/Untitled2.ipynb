```python
import sqlite3
import csv
```

```python
chadwickPath = "C:/people-0.csv"

connection = sqlite3.connect('./updated_database/chadwick.db')
cursor = connection.cursor()

with open(chadwickPath, 'r', encoding='utf-8') as csvfile:
    csv_reader = csv.reader(csvfile)
    header = next(csv_reader)

    cursor.execute(f"CREATE TABLE IF NOT EXISTS chadwick ({', '.join(header)});")

connection.commit()

cursor.close()
connection.close()
```

```python
chadwickPath = "C:/{}/chadwick_bureau_register/data/people-"

connection = sqlite3.connect('./updated_database/chadwick.db')
cursor = connection.cursor()

for i in range(10):
    concatenatedPeoplePath = chadwickPath + str(i) + ".csv"

    with open(concatenatedPeoplePath, 'r', encoding='utf-8') as csvfile:
        csv_reader = csv.reader(csvfile)
        header = next(csv_reader)

        placeholders = ', '.join(['?'] * len(header))

        for i, row in enumerate(csv_reader):
            cursor.execute(f"INSERT OR REPLACE INTO chadwick VALUES ({placeholders});", row)

connection.commit()

cursor.close()
connection.close()
```
